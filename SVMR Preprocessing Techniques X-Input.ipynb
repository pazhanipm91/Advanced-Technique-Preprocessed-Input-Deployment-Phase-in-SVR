```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#converting categorical data into numerical data
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
dependent=dataset[["Profit"]]
```

```python
independent
```

```python
dependent
```

```python
# Now We can split the training set and the test set
# We can use the same code as in the "SIMPLE LINEAR REGRESSION"
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent,dependent, test_size = 0.30,random_state = 0)

#NOTE:
# After the execution of r2_score the result was -0. THis is not near to 1. So we have to do the standardization method form here itself
```

```python
X_train
```

```python
# PRE-PROCESSING TECHNIQUES APPLIED.
# Only X-Input Going to be preprocessed and not Y-Output

#Standardization Technique: THis technique will be used only after the the result is not nearly to 1.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train) # Importance of Fit_Transform = All Data's from the X_train going to be applied in the StandardScaler formula & 
#Substituted the same data in to the (X-Tain) 
# E.g FIT()= x-mean-value/StandardDeviations = x-10/20 = -5/20.
# Fit() methods formula will be applied for the X_train Values and stored to the X_train itself and again Substitues each values from X_train and finally will transform an output.
#This routine process is called fit() Transform method.

X_test=sc.transform(X_test) #After this execution with this formula "X-Mue/sigma" again it'll save the calculation in X_train
```

```python
X_train
```

```python
#Model Creation
from sklearn.svm import SVR
regressor=SVR(kernel="sigmoid", C=3000) #These 2 lines of codes are just a procedure.

#For Non_Linear = We use the term="rbf".[Radial Bias Function]

#SVM - Used for Non-Linear Dataset, Here We do with both Linear and Non-Linear Algorithms.

#KERNEL is used to represent whether we gonna use linear/Non-Linear algorithms.
#We have to check with both the algorithms which one is giving suitable results
regressor.fit(X_train,y_train) #Training sets input and output, Hre only the MODEL gonna create.

#Here X_Trian = Preprocessed Input & y_train = Not Preprocessed.
```

```python
#Evaluation Metrics
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
